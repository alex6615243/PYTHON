```python
import pandas as pd
import tkinter as tk
from tkinter import ttk
from tkinter import filedialog

window = tk.Tk()
window.title('搜尋Excel項目')
window.geometry('600x400')

# 儲存DataFrame
df = None

# 開啟檔案
def openfile():
    global df
    file = filedialog.askopenfilename(filetypes=[('Excel files', '*.xlsx')])
    if file:
        df = pd.read_excel(file)
        df.fillna('-', inplace=True)  # 填補缺失值
        readdata(df)  # 顯示資料

# 顯示資料
def readdata(data):
    # 清除現有的資料
    for item in tree.get_children():
        tree.delete(item)
    
    # 更新Treeview的欄位
    columns = list(data.columns)
    tree["columns"] = columns
    tree["show"] = "headings"  # 隱藏第一列的標題列
    
    for col in columns:
        tree.heading(col, text=col)  # 設置每個欄位的標題
        tree.column(col, width=100, anchor='center')  # 設置每個欄位的顯示寬度和對齊方式
    
    # 將資料填入Treeview
    for i, row in data.iterrows():
        tree.insert("", "end", values=list(row))

# 搜尋資料
def search_data():
    search_term = srh_col.get().lower()  # 轉換為小寫
    if df is not None and search_term != "":
        filtered_df = df[df.apply(lambda row: row.astype(str).str.contains(search_term).any(), axis=1)]
        readdata(filtered_df)  # 顯示篩選後的資料
    elif df is not None:
        readdata(df)  # 若搜尋框為空，顯示完整資料

# UI元件佈局
btn1 = ttk.Button(window, text='選擇檔案', command=openfile)
btn1.place(x=480, y=20)

btn2 = ttk.Button(window, text='搜尋', command=search_data)
btn2.place(x=400, y=20)

srh_col = ttk.Entry(window, width=35)
srh_col.place(x=20, y=20)

# Treeview 顯示資料
tree = ttk.Treeview(window)
tree.place(x=20, y=50, width=560, height=300)

# 初始化GUI
window.mainloop()
```

```python
#各廠商期望值的計算
import numpy as np
pi_A = data['廠商A'].value_counts()/31
a1 = np.array(pi_A.index)
b1 = np.array(pi_A.values)
EX_A = int((a1*b1).sum().round())
pi_B = data['廠商B'].value_counts()/31
a2 = np.array(pi_B.index)
b2 = np.array(pi_B.values)
EX_B = int((a2*b2).sum().round())
pi_C = data['廠商C'].value_counts()/31
a3 = np.array(pi_C.index)
b3 = np.array(pi_C.values)
EX_C = int((a3*b3).sum().round())
pi_D = data['廠商D'].value_counts()/31
a4 = np.array(pi_D.index)
b4 = np.array(pi_D.values)
EX_D = int((a4*b4).sum().round())
pi_E = data['廠商E'].value_counts()/31
a5 = np.array(pi_E.index)
b5 = np.array(pi_E.values)
EX_E = int((a5*b5).sum().round())
EX_TOL = [EX_A,EX_B,EX_C,EX_D,EX_E]
print(EX_TOL)

```

```python
#各廠商的標準差
var_A = ((a1**2)*b1).sum() - EX_A**2
standard_A = np.sqrt(var_A)
var_B = ((a2**2)*b2).sum() - EX_B**2
standard_B = np.sqrt(var_B)
var_C = ((a3**2)*b3).sum() - EX_C**2
standard_C = np.sqrt(var_C)
var_D = ((a4**2)*b4).sum() - EX_D**2
standard_D = np.sqrt(var_D)
var_E = ((a5**2)*b5).sum() - EX_E**2
standard_E = np.sqrt(var_E)
STAND_TOL = [standard_A,standard_B,standard_C,standard_D,standard_E]
print(STAND_TOL)
```

```python
import pandas as pd
df = pd.DataFrame({'廠商':['A','B','C','D','E'],'期望值':EX_TOL,'標準差':STAND_TOL})
df['期望值上限'] = (df['期望值'] + df['標準差']).round()
df['期望值下限'] = (df['期望值'] - df['標準差']).round()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_excel(r"C:\Users\234500\Desktop\result.xlsx")
value_range = df['期望值上限'] - df['期望值下限']
sns.set_style('whitegrid')
sns.set_palette('Set2')

plt.figure(figsize=(10,6))
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei']
plt.rcParams['axes.unicode_minus'] = False
plt.title('一月份各廠商到工人數期望值和標準差')
plt.xlabel('廠商')
plt.ylabel('出工人數')

for idx,i  in df.iterrows():
    plt.bar(i['廠商'], value_range[idx], bottom=i['期望值下限'],width=0.5 )
    plt.text(i['廠商'],i['期望值上限'],f'{i["期望值上限"]}',ha='center',va='bottom')
    plt.text(i['廠商'],i['期望值下限'],f'{i["期望值下限"]}',ha='center',va='bottom')
    plt.text(i['廠商'],i['期望值'],f'{i["期望值"]}',ha='center',va='bottom',color='red')

    plt.hlines(y=i['期望值'], xmin=idx-0.2, xmax=idx+0.2, color='red', linestyles='--', lw=2)
sns.despine()
plt.tight_layout()
   
plt.show()
```
