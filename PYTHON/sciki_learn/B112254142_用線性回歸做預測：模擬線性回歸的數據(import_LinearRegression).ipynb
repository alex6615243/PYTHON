# CODE

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,5,50)
y = 1.2*x + 0.8 + 2 *np.random.randn(50)
X = x.reshape(50,1)
from sklearn.linear_model import LinearRegression
regr = LinearRegression()
regr.fit(X,y)
Y = regr.predict(X)
plt.scatter(x,y)
plt.plot(x,Y,'g')
```

# 作業：溫度和銷售料的線性回歸；當溫度在42度時的預測銷售量

```python
import numpy as np
import matplotlib.pyplot as plt
tem = np.array([29,28,34,31,25,29,32,31,24,33,25,31,26,30])
sales = np.array([77,62,93,84,59,64,80,75,58,91,51,73,65,84])
TEM = tem.reshape(14,1)
from sklearn.linear_model import LinearRegression
regr = LinearRegression()
regr.fit(TEM,sales)
Y = regr.predict(TEM)
plt.scatter(tem,sales)
plt.plot(tem,Y,'g')
plt.scatter(42,regr.predict([[42]]),s=100,marker='^',c='r')
```
